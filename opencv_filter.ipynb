```python
img = cv2.imread('./images/lena.jpg', 1)
```

```python
type(img)
```

```python
plt.imshow(img)
plt.title('BGR')
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('RGB')
plt.show()
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2GRAY), 'gray')
plt.title('GrayScale')
plt.show()
```

```python
threshold = 127
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, binImage = cv2.threshold(img2, threshold, 255, cv2.THRESH_BINARY)
fig, sp = plt.subplots(1, 2, sharey=True, sharex=True, figsize=(10, 5))
sp[0].imshow(img2, 'gray')
sp[0].set_title('GrayScale Image')
sp[1].imshow(binImage, 'gray')
sp[1].set_title('Binarization Image')

for i in sp:
    i.set(adjustable='box-forced',aspect='equal')
    i.get_xaxis().set_visible(False)
    i.get_yaxis().set_visible(False)

plt.show()
```

```python
# blur
Ib = cv2.blur(img2, ksize=(5,5))
# Gaussian
Ig = cv2.GaussianBlur(img2, ksize=(5,5), sigmaX=2)
# median
Im = cv2.medianBlur(img2, ksize=5)
# bilateral
Ibi = cv2.bilateralFilter(img2, d=5, sigmaColor=5, sigmaSpace=2)

fig, sp = plt.subplots(3, 2, sharey=True, sharex =True, figsize=(10,15))
sp[0][0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
sp[0][0].set_title('Original Image')
sp[0][1].imshow(img2, 'gray')
sp[0][1].set_title('GrayScale Image')
sp[1][0].imshow(Ib, 'gray')
sp[1][0].set_title('blur')
sp[1][1].imshow(Ig, 'gray')
sp[1][1].set_title('Gaussian')
sp[2][0].imshow(Im, 'gray')
sp[2][0].set_title('median')
sp[2][1].imshow(Ibi, 'gray')
sp[2][1].set_title('bilateral')

for i in sp:
    for j in i:
        j.set(adjustable='box-forced',aspect='equal')
        j.get_xaxis().set_visible(False)
        j.get_yaxis().set_visible(False)

plt.show()
```

```python
gaussian1 = cv2.GaussianBlur(img2, (7, 7), 0)
gaussian2 = cv2.GaussianBlur(img2, (7, 7), 1)
gaussian3 = cv2.GaussianBlur(img2, (7, 7), 7)

fig, sp = plt.subplots(2, 2, sharey=True, sharex =True, figsize=(10, 10))
sp[0][0].imshow(img2, 'gray')
sp[0][0].set_title('GrayScale Image')
sp[0][1].imshow(gaussian1, 'gray')
sp[0][1].set_title('Gaussian sigma = 0')
sp[1][0].imshow(gaussian2, 'gray')
sp[1][0].set_title('Gaussian sigma = 1')
sp[1][1].imshow(gaussian3, 'gray')
sp[1][1].set_title('Gaussian sigma = 7')

for i in sp:
    for j in i:
        j.set(adjustable='box-forced',aspect='equal')
        j.get_xaxis().set_visible(False)
        j.get_yaxis().set_visible(False)

plt.show()
```

```python
gaussian4 = cv2.GaussianBlur(img2, (7, 7), sigmaX=7, sigmaY=7)
gaussian5 = cv2.GaussianBlur(img2, (7, 7), sigmaX=7, sigmaY=1)
gaussian6 = cv2.GaussianBlur(img2, (7, 7), sigmaX=7, sigmaY=1)

fig, sp = plt.subplots(2, 2, sharey=True, sharex =True, figsize=(10, 10))
sp[0][0].imshow(img2, 'gray')
sp[0][0].set_title('GrayScale Image')
sp[0][1].imshow(gaussian4, 'gray')
sp[0][1].set_title('Gaussian sigmax = 7 sigmay = 7')
sp[1][0].imshow(gaussian5, 'gray')
sp[1][0].set_title('Gaussian sigma = 1 sigmay = 7')
sp[1][1].imshow(gaussian6, 'gray')
sp[1][1].set_title('Gaussian sigma = 7 sigmay = 1')

for i in sp:
    for j in i:
        j.set(adjustable='box-forced',aspect='equal')
        j.get_xaxis().set_visible(False)
        j.get_yaxis().set_visible(False)

plt.show()
```

```python
# Canny filter
edge = cv2.Canny(img2, 100, 200)

fig, sp = plt.subplots(1, 2, sharey=True, sharex =True, figsize=(9, 4.5))
sp[0].imshow(img2, 'gray')
sp[0].set_title('GrayScale Image')
sp[1].imshow(edge, 'gray')
sp[1].set_title('CANNfifY Image')

for i in sp:
    i.set(adjustable='box-forced',aspect='equal')
    i.get_xaxis().set_visible(False)
    i.get_yaxis().set_visible(False)
    
plt.show()
```
